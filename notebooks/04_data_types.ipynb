# Milestone 4.14 — Understanding Python Numeric and String Data Types

In this notebook, we explore the specific data types found within the BheedRadar footfall dataset. Understanding types like `int`, `float`, and `str` is crucial for correct data cleaning and analysis.

## 1. Load the Dataset
We use the `sample_footfall.csv` to inspect the types.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('../data/raw/sample_footfall.csv')
df.head()
```

## 2. Integer (`int`)
Numeric values without decimals. In our dataset, `visitor_count` and `is_holiday` are integers.

```python
visitor_count_sample = df['visitor_count'].iloc[0]
print(f"Value: {visitor_count_sample}")
print(f"Type: {type(visitor_count_sample)}")
```

## 3. Floating Point (`float`)
Numeric values with decimals. `temperature` is a float.

```python
temp_sample = df['temperature'].iloc[0]
print(f"Value: {temp_sample}")
print(f"Type: {type(temp_sample)}")
```

## 4. String (`str`)
Text data, typically used for categories. `city` and `zone` are strings (represented as `object` in Pandas).

```python
city_sample = df['city'].iloc[0]
print(f"Value: {city_sample}")
print(f"Type: {type(city_sample)}")
```

## 5. Type Casting
Demonstrating how to convert types (e.g., integer to float).

```python
visitor_float = float(visitor_count_sample)
print(f"Casted value: {visitor_float}")
print(f"New Type: {type(visitor_float)}")
```
