# Milestone 4.27 & 4.28 — Pandas Series, DataFrames, Indexing, and Groupby Basics

In this notebook, we move from NumPy to Pandas to handle structured data. Pandas is built on top of NumPy and provides much higher-level data structures like Series and DataFrames.

## 1. Import Pandas
Conventionally imported as `pd`.

```python
import pandas as pd
import numpy as np
```

## 2. Pandas Series
A one-dimensional array-like object that can hold any data type with an associated index.

```python
# Creating a Series from a list
visitors = pd.Series([1200, 2500, 800], index=['Jaipur', 'Goa', 'Shimla'])
print("Tourism Series:")
print(visitors)
print(f"\nIndex: {visitors.index}")
print(f"Values: {visitors.values}")
```

## 3. Pandas DataFrames
A two-dimensional, size-mutable, potentially heterogeneous tabular data structure.

```python
# Creating a DataFrame from a dictionary
data = {
    'City': ['Jaipur', 'Goa', 'Shimla', 'Jaipur', 'Goa', 'Shimla'],
    'Month': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb'],
    'Visitors': [1200, 2500, 800, 1350, 2200, 950],
    'Temperature': [20, 28, 5, 22, 30, 8]
}

df = pd.DataFrame(data)
print("Tourism DataFrame:")
print(df)
```

## 4. Indexing and Slicing (loc and iloc)
Different ways to access rows and columns.

```python
print("--- Row 0 using .iloc (Position-based) ---")
print(df.iloc[0])

print("\n--- Selecting specific columns ---")
print(df[['City', 'Visitors']])

print("\n--- Filtering for visitors > 1500 ---")
print(df[df['Visitors'] > 1500])
```

## 5. Groupby Basics
Calculate statistics for each group (e.g., Average visitors per month).

```python
print("--- Average Visitors by City ---")
city_avgs = df.groupby('City')['Visitors'].mean()
print(city_avgs)

print("\n--- Total Visitors by Month ---")
monthly_totals = df.groupby('Month')['Visitors'].sum()
print(monthly_totals)
```
