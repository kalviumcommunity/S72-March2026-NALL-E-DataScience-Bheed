# Milestone 4.38 & 4.39 — Cross-column Analysis and Distribution Plots

In this notebook, we perform deeper aggregations and use Matplotlib and Seaborn to visualize tourist footfall trends.

## 1. Import Libraries
We add `matplotlib` and `seaborn` for data visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the aesthetic style for the plots
sns.set_theme(style="whitegrid")
```

## 2. Load Data
Using the dummy dataset created in the previous milestone.

```python
df = pd.read_csv('../data/raw/tourist_data.csv')
df.head()
```

## 3. Cross-column Analysis: Temperature vs. Visitors
Let's see if there is any relationship between monthly temperature and footfall.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='temperature', y='visitor_count', hue='city', style='is_holiday', s=100)
plt.title('Temperature vs. Visitor Count by City')
plt.xlabel('Average Temperature (°C)')
plt.ylabel('Visitor Count')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## 4. Distribution Plots: Average Visitors per City
Visualizing which cities are the most popular.

```python
city_avg = df.groupby('city')['visitor_count'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=city_avg, x='city', y='visitor_count', palette='viridis')
plt.title('Average Visitor Count per City')
plt.ylabel('Average Visitors')
plt.show()
```

## 5. Holiday Impact Analysis
Understanding how public holidays affect tourism numbers.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='is_holiday', y='visitor_count')
plt.title('Impact of Holidays on Visitor Counts')
plt.xlabel('Is Holiday')
plt.ylabel('Visitor Count')
plt.show()
```
