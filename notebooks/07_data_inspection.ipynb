# Milestone 4.29, 4.30 & 4.37 — CSV Loading, Data Inspection, and Summary Statistics

In this notebook, we perform the initial steps of the Data Science life cycle on a realistic tourist footfall dataset.

## 1. Load Data from CSV
We use `pd.read_csv()` to import our raw tourist data.

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('../data/raw/tourist_data.csv')

# Display the first few rows
df.head()
```

## 2. Basic Data Inspection
Check the structure, data types, and dimensions of the dataset.

```python
print("--- Dataset Dimensions ---")
print(df.shape)

print("\n--- Column Data Types ---")
print(df.dtypes)

print("\n--- General Info ---")
df.info()
```

## 3. Summary Statistics for Numerical Columns
Understand the central tendency and spread of our quantitative variables like visitor count and temperature.

```python
print("--- Summary Statistics ---")
print(df.describe())
```

## 4. Value Counts for Categorical Columns
See how data is distributed across different cities and zones.

```python
print("--- Records per City ---")
print(df['city'].value_counts())

print("\n--- Records per Zone ---")
print(df['zone'].value_counts())
```
