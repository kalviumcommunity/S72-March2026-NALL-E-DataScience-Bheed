# Milestone 4.22 & 4.23 — NumPy Basics and Array Operations

In this notebook, we transition from standard Python lists to NumPy arrays manually and explore their core properties like shape, indexing, and slicing.

## 1. Import NumPy
Conventionally imported as `np`.

```python
import numpy as np
import time
```

## 2. Converting Python Lists to NumPy Arrays
We use sample monthly visitor data for three major cities.

```python
# Python Lists
jaipur_visitors = [1200, 1350, 1100, 900, 750, 600]
goa_visitors = [2500, 2200, 1800, 1500, 1200, 1000]
shimla_visitors = [800, 950, 1200, 1400, 1600, 1800]

# Convert to NumPy Arrays
jaipur_arr = np.array(jaipur_visitors)
goa_arr = np.array(goa_visitors)
shimla_arr = np.array(shimla_visitors)

print(f"Jaipur Array: {jaipur_arr}")
print(f"Type: {type(jaipur_arr)}")
```

## 3. Performance Benchmark: Python Lists vs NumPy Arrays
Demonstrating why we use NumPy for large-scale data.

```python
large_list = list(range(1000000))
large_arr = np.array(large_list)

# Benchmark Python List Sum
start_time = time.time()
sum_list = sum(large_list)
print(f"Python List Sum Time: {time.time() - start_time:.6f} seconds")

# Benchmark NumPy Array Sum
start_time = time.time()
sum_arr = np.sum(large_arr)
print(f"NumPy Array Sum Time: {time.time() - start_time:.6f} seconds")
```

## 4. Array Shape, Dimensions, and Size
Essential attributes for multi-city analysis.

```python
print(f"Jaipur Shape: {jaipur_arr.shape}")
print(f"Dimensions: {jaipur_arr.ndim}")
print(f"Size (Total elements): {jaipur_arr.size}")
```

## 5. Indexing and Slicing
Accessing specific months of tourism data.

```python
# Get visitor count for the 3rd month (March)
march_jaipur = jaipur_arr[2]
print(f"March visitors (index 2): {march_jaipur}")

# Get visitors for the first quarter (Months 0, 1, 2)
q1_jaipur = jaipur_arr[0:3]
print(f"Q1 Jaipur Visitors: {q1_jaipur}")

# Get the last two months recorded
last_two = jaipur_arr[-2:]
print(f"Last two months: {last_two}")
```

## 6. Multi-City Matrix (2D Array)
Combining data into a matrix representation.

```python
tourism_matrix = np.array([jaipur_arr, goa_arr, shimla_arr])
print("Tourism Matrix (Cities x Months):")
print(tourism_matrix)
print(f"Matrix Shape: {tourism_matrix.shape}") # Output: (3, 6)
```
