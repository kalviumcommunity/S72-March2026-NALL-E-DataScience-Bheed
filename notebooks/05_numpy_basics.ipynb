# Milestone 4.22 — Creating NumPy Arrays from Python Lists

In this notebook, we transition from standard Python lists to NumPy arrays. NumPy provides high-performance multidimensional array objects and tools for working with these arrays, which is essential for BheedRadar's large-scale tourism forecasting.

## 1. Import NumPy
Conventionally imported as `np`.

```python
import numpy as np
import time
```

## 2. Converting Python Lists to NumPy Arrays
We use sample monthly visitor data for three major cities.

```python
# Python Lists
jaipur_visitors = [1200, 1350, 1100, 900, 750, 600]
goa_visitors = [2500, 2200, 1800, 1500, 1200, 1000]
shimla_visitors = [800, 950, 1200, 1400, 1600, 1800]

# Convert to NumPy Arrays
jaipur_arr = np.array(jaipur_visitors)
goa_arr = np.array(goa_visitors)
shimla_arr = np.array(shimla_visitors)

print(f"Jaipur Array: {jaipur_arr}")
print(f"Type: {type(jaipur_arr)}")
```

## 3. Performance Benchmark: Python Lists vs NumPy Arrays
Demonstrating why we use NumPy for large-scale data.

```python
large_list = list(range(1000000))
large_arr = np.array(large_list)

# Benchmark Python List Sum
start_time = time.time()
sum_list = sum(large_list)
print(f"Python List Sum Time: {time.time() - start_time:.6f} seconds")

# Benchmark NumPy Array Sum
start_time = time.time()
sum_arr = np.sum(large_arr)
print(f"NumPy Array Sum Time: {time.time() - start_time:.6f} seconds")
```
