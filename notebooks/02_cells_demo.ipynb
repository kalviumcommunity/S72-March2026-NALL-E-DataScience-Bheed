# Milestone 4.8 — Understanding Notebook Cells: Code vs Markdown

In this notebook, we demonstrate the difference between **Code** cells and **Markdown** cells using a tourism footfall example.

## 1. Code Cells
Code cells are used to execute Python logic. Below, we load a small sample of monthly visitor data for Jaipur and compute the average footfall.

```python
# Sample data: Monthly visitors for Jaipur (Jan-Dec)
jaipur_visitors = [1200, 1350, 1100, 900, 750, 600, 850, 950, 1050, 1400, 1600, 1550]

average_footfall = sum(jaipur_visitors) / len(jaipur_visitors)
print(f"Average monthly footfall in Jaipur: {average_footfall:.2f}")
```

## 2. Markdown Cells
Markdown cells are used for documentation, explanation, and visual formatting. 

### Formatting Techniques:
- **Bold Text**: `**Text**` -> **BheedRadar**
- *Italic Text*: `*Text*` -> *Forecasting*
- [x] Task Lists
- LaTeX for math: $E = mc^2$ or $\bar{x} = \frac{\sum x_i}{n}$
